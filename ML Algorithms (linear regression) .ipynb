```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [3,21,22,34,54,34,55,67,89,99]
x = np.array(x)
y = [1,10,14,34,44,36,22,67,79,90]
y =np.array(y)

plt.scatter(x,y)
```

```python
# the basic linear regression model is wx+b.
def model(a,b,x):
    return a*x + b
```

```python
#the most commonly used loss function of kinear regression.
def loss_function(a,b,x,y):
    num = len(x)
    prediction = model(a,b,x)
    loss = (0.5/num) * (np.square(prediction-y)*x).sum()
    return loss
```

```python
#the optimization function mainly uses partial derivatives of the loss function on a and b.
def optimize(a,b,x,y):
    num = len(x)
    prediction = model(a,b,x)
    # update the values of a and b by finding the partial derivatives of the loss function on a and b.
    da = (1.0/num) * ((prediction - y)*x).sum()
    db = (1.0/num) * ((prediction - y).sum())
    a = a - lr*da                  
    b = b - lr*db
    return a , b 
```

```python
#iterated function, return a and b.
def iterate(a,b,x,y,times):
    for i in range(times):
        a,b = optimize(a,b,x,y)
    return a,b
```

```python
#initialize parameters and display.
a=np.random.rand(1)
print(a)
b=np.random.rand(1)
print(b)
lr = 1e-4
```
